```python
import numpy as np

import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1,520000,130000,240000,110000],   #paradise biryani
    [2,120000,12000,140000,150000],
    
    [3,120000,130000,640000,150000],
    [4,320000,130000,140000,110000],
    
])


print(sales_data.shape)


print(np.sum(sales_data,axis=0))
print(np.sum(sales_data,axis=1))


yearly_total = np.sum(sales_data[:,1:],axis =0)
yearly_total = np.sum(sales_data[:,1:],axis =1)

print(yearly_total)



cumsum = np.cumsum(sales_data[:,1:],axis=1)

print(cumsum)




plt.figure(figsize=(7,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("average cumsum plot ")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid()
plt.show
```
