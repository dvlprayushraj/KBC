```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


```

```python
x = np.linspace(0,5,11)
x

y = x**2
y
```

```python

plt.title("hello")

plt.xlabel("ayush")
plt.ylabel("cachi")

plt.subplot(2,2,4)
plt.plot(x,y)
```
