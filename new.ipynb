```python
data_types = [('name', 'S15'), ('class', int), ('height', float)]

student_details = [('James', 5, 48.5), ('Nail', 6, 52.5), ('Paul', 5, 42.10), ('Pit', 5, 40.11)]

students = np.array(student_details, dtype= data_types)

print("original array ")
print(student_details)


print("arragnged array as height " )
print(np.sort(students,order='height'))
```

```python
student_id = np.array([1023, 5202, 6230, 1671, 1682, 5241, 4532])

student_height = np.array([40., 42., 45., 41., 38., 40., 42.0])


height_sorted = np.lexsort((student_id,student_height))


print(height_sorted)


for n in height_sorted:
    print(student_id[n],student_height[n])
```

```python
import numpy as np

# Get today's date with 'D' for day resolution
today = np.datetime64('today', 'D')
# Calculate yesterday's date by subtracting one day
yesterday = today - np.timedelta64(1, 'D')
# Calculate tomorrow's date by adding one day
tomorrow = today + np.timedelta64(1, 'D')

print("Yesterday:", yesterday)
print("Today:", today)
print("Tomorrow:", tomorrow)
```
