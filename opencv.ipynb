```python
import cv2

image_location = input("Enter the location of the image: ").strip()

if not image_location:
    print("No image path provided.")
else:
    image = cv2.imread(image_location)

    if image is None:
        print(f"Failed to load image from '{image_location}'. Check the path and file format.")
    else:
        grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

        user_choice = input("Do you want to save or show the image? (save/show): ").strip().lower()

        if user_choice == "save":
            cv2.imwrite("grey_image.jpg", grey)
            cv2.imshow("image", image)
            cv2.waitKey(0)
            cv2.destroyAllWindows()
            print("Image saved successfully.")
        elif user_choice == "show":
            cv2.imshow("grey", grey)
            cv2.waitKey(0)
            cv2.destroyAllWindows()
        else:
            print("Unknown choice. Please enter 'save' or 'show'.")
    

```
